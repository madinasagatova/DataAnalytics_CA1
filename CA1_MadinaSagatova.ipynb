### Normal Distribution 
Normal Distribution or Gaussian Distribution is known as a bell-shaped symmetric probability distribution. Many things are actually normally distributed, or very close to it such as height or IQ score. The Empirical Rule also called 66-95-99.7 Rule says: In a normal distribution, approximately 66% of the data falls within one standard deviation of the mean, 95% falls within two standard deviations, and 99.7% falls within three standard deviations. 
- We perform The Shapiro-Wilk test for each county and choose one which is close to normal distribution besed on the The Shapiro-Wilk test results
- Obtain z-score and Probability (Number of Trips >700.000) for county Wexford
- Obtain z-score and Probability (Number of Trips < 500.000) for county Wexford
- And obtain probability (700.000 < Number of Trips < 500.000) for county Wexford 
- We visually check if a numeric column follows a normal distribution by plotting a histogram and comparing it to a normal distribution curve.

```python
from scipy.stats import kstest
#perform The Shapiro-Wilk test - widely used goodness-of-fit test for normality
wexford_test = stats.shapiro(wexford_df['Number_Of_Trips'])
cork_test = stats.shapiro(cork_df['Number_Of_Trips'])
dublin_test = stats.shapiro(dublin_df['Number_Of_Trips'])
kerry_test = stats.shapiro(kerry_df['Number_Of_Trips'])
galway_test = stats.shapiro(galway_df['Number_Of_Trips'])
total_test = stats.shapiro(total_df['Number_Of_Trips'])
print(f"Wexford: {wexford_test}")
print(f"Cork: {cork_test}")
print(f"Dublin: {dublin_test}")
print(f"Kerry: {kerry_test}")
print(f"Galway: {galway_test}")
print(f"Total number of trips: {total_test}")
```

**Observations**
- After performing the Shapiro-Wilk test we can see that the most bigger P value have *wexford_test* and *kerry_test*. The P values are 0.94 and 0.86 respectively, which is more likely normally distributed.

```python
# calculate the mean and the standard deviation for wexford_df 
mean_wexford = wexford_df['Number_Of_Trips'].mean()
std_wexford = wexford_df['Number_Of_Trips'].std()

#print mean and std
print(f'Mean of "Number_Of_Trips" in county Wexford: {mean_wexford}')
print(f'Standart Deviation of "Number_Of_Trips" in county Wexford: {std_wexford}')
```

```python
#what is the probability of Number of Trips higher than 700.000 in Wexford for a random year
#first calculate z-score 
z_high_number = (700000 - mean_wexford)/std_wexford

#calculate probability of getting higher than 700.000
#to find the shaded area you take away probability from 1 - the total area under the curve
probability_high = 1 - norm.cdf(z_high_number)

#print z-value and probability
print(f"Z-score {z_high_number}")
print(f"Probability {probability_high}")
```

```python
#what is the probability of Number of Trips lower than 500.000 in Wexford for a random year
#first calculate z-score 
z_low_number = (500000 - mean_wexford)/std_wexford

#calculate probability of getting lower than 500.000
probability_low = norm.cdf(z_low_number)

#print z-value and probability
print(f"Z-score {z_low_number}")
print(f"Probability {probability_low}")
```

```python
#what is the probability of Number of Trips between 500.000 and 700.000 in Wexford for a random year
#calculate probability 
probability = norm.cdf(z_high_number) - norm.cdf(z_low_number)

#print probability
print(f"Probability {probability}")
```

**Observations**
- Based on the result the number of trips tends to be more likely to exceed 700.000 than fall below 500.000. And there is a relatively high probability (0.48) of the number of trips falling between 500.000 and 700.000. We can conclude that the distribution of the number of trips in county Wexford is more likely to follow a normal distribution.

### Poisson distribution 
The Poisson Distribution is a discrete probability distribution. It describes the number of events that occur in a fixed interval of time or space, given a known average rate of occurrence (assuming the events happened independently of each other). λ (lambda) is a parameter of the Poisson Distribution. It is the average rate of occurrence of the events within an interval. 
Apply the Poisson Distribution to model the number of occurrences of every county in the fixed number of samples:
- First calculate the λ - average rate: the average rate of occurrence for each county
- Apply the Poisson Distribution: use the Poisson distribution to model the number of occurrences in a fixed number of samples
- Plot the PMF(Probability Mass Function) of the Poisson Distribution for each county to visualize the probability of observing different numbers of occurrence

```python
#calculate the number of occurrences of each country in the Data Set 
counties_count = counties_trips_df['Counties'].value_counts()

#set number of trials
num_samples = 15

#print the number of occurrences
counties_count
```

```python
#empty dictionary for storing Poisson distribution for each county 
poisson_distributions = {}

#applying Poisson distribution for each county
for counties, count in counties_count.items():
    rate = count/num_samples #calculate average occurrence per every sample  
    poisson_distribution = poisson(mu=rate)
    poisson_distributions[counties] = poisson_distribution
```

```python
#possible number of occurrences(0 to num_samples) 
k_values = np.arange(num_samples + 1)

#plot Poisson distribution 
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
for counties,poisson_distribution in poisson_distributions.items():
    pmf_values = poisson_distribution.pmf(k_values) #PMF 
    plt.plot(k_values,pmf_values, marker='o', linestyle='-', label = counties)
    
plt.title('Poisson Distribution for Observing Each County in 15 Samples')
plt.xlabel('Number of Observations')
plt.ylabel('Probability of Successes')
plt.xticks(k_values)
plt.grid(True)
#save as a .png 
plt.savefig('poissonDitribution.png')
plt.show()
```

```python

X = np.array(total_df.index).reshape(-1,1)
y = total_df['Number_Of_Trips']

model = LinearRegression()
model.fit(X,y)

prospective_years = np.array(range(max(total_df.index) + 1, max(total_df.index) + 10)).reshape(-1,1)
prospective_trips = model.predict(prospective_years)

plt.figure(figsize=(10,8))
plt.plot(total_df.index, total_df['Number_Of_Trips'], label = 'Actual Total Domestic Number of Trips', marker = 'o')
plt.plot(prospective_years, prospective_trips, label = 'Forecasted Total Domestic Number of Trips', marker = 'o', c='green')

sns.set(style="white") #style 
plt.xlabel("Year", size = 10)
plt.ylabel("Total Number of Trips", size = 10)
plt.title('Growth of Domestic Trips in Ireland Forecast: 2023 to 2030 (Linear Regression)')
plt.legend()
plt.grid(False)
#save as a .png 
plt.savefig('tourismForecastLinearRegression.png')
plt.show()
```

```python
#predict the target variable for the features 
y_predict = model.predict(X)

#calculating Mean Squared Error
mse = mean_squared_error(y, y_predict)

#calculating Root Mean Square Error
rmse = np.sqrt(mse)

#calculating R-squared Error
r2 = r2_score(y, y_predict)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Square Error: {rmse:.2f}" )
print(f"R-squared Error: {r2:.2f}")
```
